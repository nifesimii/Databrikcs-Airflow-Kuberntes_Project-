```python
stg_posts_df = spark.table("default.stg_posts")
```

```python
display(stg_posts_df)
```

```python
from pyspark.sql import DataFrame
import pyspark.sql.functions as F

def posts_top_tags(stg_posts_df: DataFrame) -> DataFrame:
    return (
        stg_posts_df
        .withColumn("tag_exploded", F.explode("TagsArray"))
        .groupBy("tag_exploded").agg(F.approx_count_distinct("PostId").alias
        ("tags_count"))
        .orderBy(F.col("tags_count").desc())
        )
    
marts_top_tag_df = posts_top_tags(stg_posts_df)
marts_top_tag_df.write.mode("overwrite").saveAsTable("marts_top_tags")
```
